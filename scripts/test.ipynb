```python
import httpx


def info_tool(city: str, natural_factor: str, start_date: str, end_date: str):
    query_num = get_query_num(city, natural_factor)
    start_timestamp = convert(start_date)
    end_timestamp = convert(end_date)

    url = f"https://visualize.openiot.in/api/ds/query?ds_type=influxdb&requestId=Q{query_num}"

    headers = {
        "Accept": "application/json, text/plain, */*",
        "Accept-Encoding": "gzip, deflate, br, zstd",
        "Accept-Language": "en-US,en;q=0.9",
        "Host": "visualize.openiot.in",
        "Origin": "https://visualize.openiot.in",
        "Referer": "https://visualize.openiot.in/d/OBBdNCmVzwsdsd/aletty-micro-climate-monitoring?orgId=2&from=now%2Ffy&to=now%2Ffy",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/139.0.0.0 Safari/537.36",
        "content-type": "application/json",
        "sec-ch-ua": '"Not;A=Brand";v="99", "Google Chrome";v="139", "Chromium";v="139"',
        "sec-ch-ua-mobile": "?0",
        "sec-ch-ua-platform": '"Windows"',
        "x-dashboard-uid": "OBBdNCmVzwsdsd",
        "x-datasource-uid": "denukohoa42kgf",
        "x-grafana-device-id": "9ce10ff066bc28345d129ee3fb6c1018",
        "x-grafana-org-id": "2",
        "x-panel-id": "22",
        "x-panel-plugin-id": "stat",
        "x-plugin-id": "influxdb",
    }

    # Map natural factors to InfluxDB measurement names
    measurement_map = {
        "humidity": "device_frmpayload_data_humidity",
        "temperature": "device_frmpayload_data_temperature",
        "rainfall": "device_frmpayload_data_rainfall",
        "pressure": "device_frmpayload_data_pressure",
    }

    measurement = measurement_map.get(natural_factor.lower())
    if not measurement:
        raise ValueError(f"Invalid natural factor: {natural_factor}")

    # Device name logic (could be dynamic if needed)
    device_name = "ICF_THS05_005"

    # Flux query (InfluxDB)
    flux_query = f"""
        from(bucket: "nemom")
          |> range(start: v.timeRangeStart, stop: v.timeRangeStop)
          |> filter(fn: (r) => r["device_name"] == "{device_name}")
          |> filter(fn: (r) => r["_measurement"] == "{measurement}")
          |> filter(fn: (r) => r["_field"] == "value")
          |> yield(name: "last")
    """

    payload = {
        "queries": [
            {
                "datasource": {"type": "influxdb", "uid": "denukohoa42kgf"},
                "groupBy": [],
                "orderByTime": "ASC",
                "policy": "default",
                "query": flux_query,
                "refId": "A",
                "resultFormat": "time_series",
                "select": [
                    [
                        {"params": ["value"], "type": "field"},
                        {"params": [], "type": "sum"},
                    ]
                ],
                "tags": [{"key": "device_name", "operator": "=", "value": device_name}],
                "datasourceId": 42,
                "intervalMs": 200,
                "maxDataPoints": 99999999,
            }
        ],
        "from": start_timestamp,
        "to": end_timestamp,
    }

    response = httpx.post(url, headers=headers, json=payload)
    return response.json()
```
